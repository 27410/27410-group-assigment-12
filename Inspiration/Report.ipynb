# Construction of GSM for *Vibrio natriegens* and production of beta carotene

## 1. Introduction

### 1.1 Literature review of the cell factory (*V. natriegens*)

The fast-growing marine *Vibrio natriegens* is an emerging synthetic biology chassis and a promising candidate for an industrial cell factory. The bacterium possesses many desirable qualities for a cell factory making it interesting for industrial application. This includes a high growth and substrate uptake rate, metabolic diversity, genetic tractability, and tolerance against high salinity (Hoff et al. 2020). These properties could remarkably decrease production expenses and thus increase profits. *V. natriegens* belongs to the class of gammaproteobacteria, just like *E.coli*, and provides a very promising alternative to this commonly used industrial workhorse. 

The high growth rate of *V. natriegens* makes it highly interesting as a commercial cell factory. *V. natriegens* is reported to have the highest growth rate hitherto observed in a non-pathogenetic organism and has a generation time of less than 10 minutes (Hoffart et al. 2017). In comparison, the growth rate for *E. coli* is usually 15-20 minutes. This trait could prove very timesaving and speed up the Design, Build, Test, and Learn (DBTL) cycle for the cell factory development (Eichmann et al. 2019).

The metabolic diversity and requirement of high salinity of *V. natriegens* offers a remarkable potential to decrease the production costs. In industrial fermentation, the cost of water is a significant expense. *V. natriegens* is ubiquitous in the marine environment and thus thrive in high salinity. This means that seawater can be used in the fermentation process for *V. natriegens* and thus significantly reducing the expense of water. Furthermore, the marine bacterium has a high metabolic diversity and can utilize a wide range of substrates including cheap substrates such as sucrose and glycerol (Hoffart et al. 2017; Ellis et al. 2019). Using *V. natriegens* as a cell factory could thus prove very economically favorable.

There are currently several established molecular biology techniques for *V. natriegens* including electroporation, conjugation, and CRIPSPR-Cas9 (Hoff et al. 2020). Recent studies have revealed that many methods and tools for production in *E. coli* can be applied in *V. natriegens* (Tschirhart et al. 2019; Eichmann et al. 2019). The codon usage in the two organisms are fairly similar (Hoff et al. 2020). A recent study investigated how well genetic components commonly used for *E. coli* work in *V. natriegens* (Tschirhart et al. 2019). This revealed that many of the synthetic biology tools can also be applied in the fast-growing vibrio. It has been suggested that *E. coli* tools and protocols could be a good starting point for developing more tools for *V. natriegens* (Hoff et al. 2020). This signifies that tools and methods for *V. natriegens* can hopefully be expeditiously developed for the optimization of *V. natriegens* as a cell factory.


### 1.2 Literature review of the compound (beta carotene)

The natural product beta carotene is terpenoid that are naturally produced in fungi, plant and fruits. It is a colored red-orange pigment by which is it often used for food coloring in the food and beverage industry. It is also used for natural cosmetic products and besides being a pigment, it is also the most common source for provitamin A. The production of vitamin A is of interesting because vitamin A has multiple functions in the body, such as vision and the immune system. To express beta carotene in *V. natriegens*, the beta carotene pathway from Vibrio campbellii and the mevalonate (MVA) biosynthetic pathway from the Gram-positive bacterium *Lactobacillus acidophilus* were introduced to get the right precursors. 
Three pathways are relevant for beta carotene production (Figure 1). *Vibrio natriegens* already produce the important precursors isopentenyl pyrophosphate (IPP) and dimethylallyl pyrophosphate (DMAPP) through its upstream methylerythritol 4-phosphate (MEP) pathway, but this product seems to be a bottleneck in the production of beta carotene. To solve this issue, the MVA pathway from *Lactobacillus acidophilus* is introduced and thereby more IPP is produced. IPP is then used in the beta carotene pathway from *Vibrio campbellii* to reach the final product. Beta carotene have already been experimentally heterologously expressed in *V. natriegens* and other microbial chassis with great success. By making a computed *V. natriegens* cell factory the goal is to increase the production rate of beta-carotene by model optimizing (Ellis et al. 2019).

## 2. Problem definition

*V. natriegens* is an emerging cell factory with many attractive properties – most notable, its incredibly high growth rate. Furthermore, it can be cultivated using cheap substrates and salt water.  The latter could reduce the use of drinking water significantly and as clean drinking water is a limiting resource it can help the UN with sustainable development goal 6. *V. natriegens* has been proposed as a promising alternative to *E. coli*. However, less information and computational tools are currently available for *V. natriegens* compared to *E. coli*. It is thus highly relevant to construct a genome-scale model of this organism in order to e.g. computationally perform cell factory engineering for future use of *V. natriegens* as a cell factory. 

This project will produce the high demanded product beta carotene in a more efficient and environmentally friendly way. Beta carotene have been used as a dye for both human and animal feed for many years. Today’s production of beta carotene is 95% chemical and 5% natural. This chemical produced beta carotene can be damaging to the environment and moreover the increased focus on human health and natural produced products demands for a more sustainable and natural production. It would therefore have an effect on the climate change mitigation to change a larger part of the chemical produced beta-carotene to bioproduced beta-carotene. To achieve this, the heterologous pathway for beta-carotene can be introduced into the marine bacterium *V. natriegens*.

## 4. Computer-Aided Cell Factory Engineering

The heterologous pathway for beta carotene production will be inserted into the constructed *V. natriegens* model as proof-of-concept to assess the model and *V. natriegens* as an emerging cell factory. For *V. natriegens* to produce beta carotene, 9 reaction steps must be added to the model. The required reaction steps are identified using KEGG PATHWAY maps of *V. natriegens* and *V. campbellii*. The first reaction is the conversion of farnesyl disphosphate and isopentenyl diphosphate into geranylgeranyl diphosphate and diphosphate. The stochiometry is defined and the reaction is added:

**Defining and adding reaction steps for beta carotene production**

```python
from cobra import Reaction, Metabolite
```

```python
Reaction_0_beta=Reaction('Step0_beta') # Farnesyl diphosphate + isopentenyl diphosphate <=> geranylgeranyl diphosphate + diphosphate
geranylgeranyl_PP=Metabolite(id='Geranylgeranyl_pp_c', compartment='c')
Reaction_0_beta.add_metabolites({modelc.metabolites.frdp_c: -1, # Farnesyl diphosphate
                                 modelc.metabolites.ipdp_c: -1, # Isopentenyl diphosphate
                                 geranylgeranyl_PP: 1, # Geranylgeranyl diphosphate
                                 modelc.metabolites.ppi_c: 1 # diphosphate
                                })

modelc.add_reactions([Reaction_0_beta]) # Adding reaction to model
```

This procedure was repeated for the remaining reactions in file Beta+MVA-pathway.ipynb, To get a flux, a boundary was implemented for beta carotene. This resulted in modelBC.xml that has the beta carotene pathway incorporated.

```python
modelBC = read_sbml_model('./Models/modelBC.xml')
```

**Predicting maximal theoretical beta carotene production**

The theoretical maximal yield of beta carotene is subsequently calculated. This is done by setting the final reaction, yielding beta carotene, as the model objective and thus maximize the flux through that reaction in a flux balance analysis. The maximal theoretical yield of beta carotene will then be the objective value of the solution. The growth rate,  beta carotene production and yield on glucose is calculated to evaluate the heterologous expression of beta carotene.

```python
with modelBC:
    modelBC.objective = modelBC.reactions.Growth
    growth_rate = modelBC.optimize().objective_value
    modelBC.objective = modelBC.reactions.Step8_beta
    beta_carotene_production = modelBC.optimize().objective_value
    beta_carotene_glc_yield = beta_carotene_production / (-1*modelBC.reactions.EX_glc__D_e.flux)
    
print('Growth rate for V. natriegens [h^-1]:', growth_rate)
print('Max. beta-carotene production [mmol gDW^-1 h^-1]:', beta_carotene_production)
print('Theoretical maximum yield [mmol_beta / mmol_glc]:', beta_carotene_glc_yield)
```

This gives a result of the maximal theoretical production of beta carotene being 2.02 mmol gDW<sup>-1</sup>h<sup>-1</sup> in *V. natriegens*. The beta-carotene pathway was also inserted into the *E. coli* iML1515 model for comparison (See file Betacarotene_in_Ecoli.ipynb). This resulted in a maximum theoretical production of 0.94 mmol gDW<sup>-1</sup>h<sup>-1</sup>. Production in the *V. natriegens* model is thus more than 2-fold higher. The maximal growth rate of *V. natriegens* is also slightly higher than *E. coli* iML1515. The maximal growth rate of *V. natriegens* ended up being 0.92 h<sup>-1</sup>, and for *E. coli* iML1515 the maximal growth rate is 0.88 h<sup>-1</sup>. However, the yields of beta carotene in *V. natriegens* and *E. coli* are more or less the same at 0.094 mmol<sub>beta</sub>/mmol<sub>glc</sub> with the first difference on the fourth significant decimal.

**Adding MVA pathway to increase concentration of precursor**

In an attempt to increase beta carotene production, the MVA-pathway has furthermore been implemented in the model (See file Beta+MVA-pathway.ipynb). The MVA pathway produces the precursors IPP and DMAPP and it is thus expected to increase the production as observed in literature (Ellis et al. 2019). The resulting model (modelBCMVA.xml) contains both the beta carotene and MVA-pathway.

```python
modelBCMVA = read_sbml_model('./Models/modelBCMVA.xml')
```

```python
with modelBCMVA:
    modelBCMVA.objective = modelBCMVA.reactions.Growth
    growth_rate_MVA = modelBCMVA.optimize().objective_value
    modelBCMVA.objective = modelBCMVA.reactions.Step8_beta
    beta_carotene_production_MVA = modelBCMVA.optimize().objective_value
    beta_carotene_glc_yield_MVA = beta_carotene_production_MVA / (-1*modelBCMVA.reactions.EX_glc__D_e.flux)
    
print('Growth rate for V. natriegens [h^-1]:', growth_rate_MVA)
print('Max. beta-carotene production [mmol gDW^-1 h^-1]:', beta_carotene_production_MVA)
print('Theoretical maximum yield [mmol_beta / mmol_glc]:', beta_carotene_glc_yield_MVA)
```

Implementation of the MVA pathway only results in negligible changes. This is presumably due to the model not having a concept of concentrations. Increasing precursor concentration thus does not affect the beta carotene production.

**Phenotypic Phase Plans**

Phenotypic phase plans are used to assess how different factors are related such as beta carotene and biomass production. 

```python
from cobra.flux_analysis.phenotype_phase_plane import production_envelope, add_envelope
```

```python
ppp_beta = production_envelope(modelBC,
                    reactions=[modelBC.reactions.Growth],
                    objective=modelBC.reactions.Step8_beta)
ppp_beta.plot(x='Growth', y='flux_maximum')
```

The maximum flux of beta-carotene is inverse proportional to the growth rate. It is thus important to outweigh these in order to achieve the best result for beta carotene production.

A phenotypic phase plane analysis for oxygen uptake against beta carotene production and growth is also conducted.

```python
with modelBC:
    modelBC.reactions.EX_o2_e.bounds = -50, 0
    ppp_o2beta = production_envelope(modelBC,
                        reactions=[modelBC.reactions.EX_o2_e],
                        objective=modelBC.reactions.Step8_beta)
    ppp_o2beta.plot(x='EX_o2_e', y='flux_maximum')
    print("The optimal O2 uptake for beta carotene flux is:",ppp_o2beta.EX_o2_e[ppp_o2beta.flux_maximum.idxmax()])
```

```python
with modelBC:
    modelBC.reactions.EX_o2_e.bounds = -50, 0
    ppp_o2X = production_envelope(modelBC,
                        reactions=[modelBC.reactions.EX_o2_e],
                        objective=modelBC.reactions.Growth)
    ppp_o2X.plot(x='EX_o2_e', y='flux_maximum')
    print("The optimal O2 uptake for growth is:",ppp_o2X.EX_o2_e[ppp_o2X.flux_maximum.idxmax()])
```

Oxygen uptake is required for both beta carotene production and growth. The optimum is around 16 mmol/g/h, and it may be advantagous to ensure this oxygen uptake during fermentation to optimize beta carotene production.

```python
with modelBC:
    modelBC.reactions.EX_glc__D_e.bounds = -100,0
    ppp_glcbeta = production_envelope(modelBC,
                        reactions=[modelBC.reactions.EX_glc__D_e],
                        objective=modelBC.reactions.Step8_beta)
    ppp_glcbeta.plot(x='EX_glc__D_e', y='flux_maximum')
    print("The optimal glucose uptake for beta carotene flux is:",ppp_glcbeta.EX_glc__D_e[ppp_glcbeta.flux_maximum.idxmax()])
```

The optimal glucose uptake rate is 50 mmol/g/h for beta carotene production. The glucose uptake by *V. natriegens* was measured as 28 mmol/g/h (Long et al. 2017), so *V. natriegens* should be cultivated with excess of glucose to ensure the highest beta carotene.

## 5. Discussion

The GSM model was constructed from carveme and adjusted according to experimental fluxes and uptakes found in Long *et al.* 2017. However, the growth rate obtained in this report with the newly constructed GSM model was 0.92 h<sup>-1</sup> which is significantly lower than the experimentally determined growth rate which is 1.7 h<sup>-1</sup>. It is still unknown why this is, but for further research more experimental data should be collected to adjust the GSM model even more to match the living cell. Additionally, this would likely result in an increased score of the model in the MEMOTE evaluation. Still, the consructed GSM model for *V. natriegens* yields a higher growth rate than the *E. coli* iML1515 model which had a growth rate of 0.88 h<sup>-1</sup>.

Beta-carotene has previously been successfully expressed in *V. natriegens* and thus verifies that the heterologous expression is possible (Ellis et al. 2019). In that study, the titer of beta-carotene was approximately 0.5 mg/L without adding the MVA pathway. The beta-carotene production in *V. natriegens* was increased approximately 6-fold in LBv2 rich medium when the MVA pathway from *Lactobacillus acidophilus* NCFM was additionally introduced (Ellis et al. 2019). When the MVA-pathway was introduced into *V. natriegens*, the titer was almost 3 mg/L (Ellis et al. 2019). The titer was not calculated in this report as no volume was mentioned. However, the introduction of the MVA pathway in the constructed GSM did not result in a difference in beta carotene production rate in *V. natriegens*. The reason for this could likely be that the GSM model does not take concentrations into count when calculating the fluxes of beta-carotene, and therefore no bottleneck was observed for limiting precursors, IPP and DMAPP. 

The maximal theoretical production of beta-carotene was observed to be more than 2-fold higher in *V. natriegens* compared to when introducing the beta-carotene pathway from *V. campbellii* in the *E. coli* iML1515 model. The heterologous expression of beta-carotene in *V. natriegens* in this study was meant as a proof-of-concept to investigate if production in this host was possible. The aim was thus not to improve the yield of beta-carotene. For this reason, it might be possible to experimentally achieve a higher yield.

At this point, the model had an overall score of 23 % in the MEMOTE evaluation. It would thus be highly relevant to look into ways of improving the model to make it more accurately depict *V. natriegens* for its use as a cell factory. For further research, one should look into optimizing the GSM model by incorporating more experimental data with fluxes of precursors to determine the uptake and flux rates more precisely. Also, it might be necessary to do some gene knockouts for further research to obtain the experimentally determined growth rate of *V. natriegens*. 


## 6. Conclusion

In this work, a GSM of the emerging cell factory *V. natriegens* has been constructed, validated, and applied. The GSM was validated with experimental data from literature to improve the model and its predictions. Achieving the experimental growth rate in the GSM has not been possible so far. The constructed GSM predicted a growth rate of 0.92 h<sup>-1</sup> while an
experimental study reported a growth rate of 1.70 h<sup>-1</sup>. The model was evaluated using MEMOTE which yielded an overall score of 22% to begin with. The GSM achieved an overall score of 23% After altering uptake rates, intracellular fluxes and deactivating certain pathways. Specifically, "Unbound Bound Flux In Default medium" had improved from 60.1% to 80.3%. The heterologous pathway for beta-carotene was successfully inserted and expressed in the model as proof-of-concept. The yield of beta carotene in *V. natriegens* was almost the same as *E. coli* at 0.094 mmol<sub>beta</sub>/mmol<sub>glc</sub>. The GSM predicted a maximal theoretical production of 2.02 mmol gDW<sup>-1</sup>h<sup>-1</sup> beta carotene in *V. natriegens* which is 2-fold higher than the *E. coli* iML1515 model. Furhtermore, the obtained growth rate of *V. natriegens* was slightly higher than for *E. coli*. This indicates that *V. natriegens* could be a favorable host for heterologous expression of beta carotene.
